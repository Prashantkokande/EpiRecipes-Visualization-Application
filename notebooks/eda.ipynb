#### What are the most common ingredients in highly rated recipes?

```python
from collections import Counter

highly_rated = result[result['rating'] >= 4.5]

highly_rated['ingredients'] = highly_rated['ingredients'].apply(lambda x: eval(x) if isinstance(x, str) else x) #Eval function evaluate python expression which are written as strings.

all_ingredients = [ingredient for sublist in highly_rated['ingredients'] for ingredient in sublist] # List comprehension

ingredient_counts = Counter(all_ingredients) # Counter takes a list as input and returns a dictionary-like object where keys are the list elements and values are their counts.

most_common_ingredients = ingredient_counts.most_common(10) 

print("Most Common Ingredients in Highly Rated Recipes:")
for ingredient, count in most_common_ingredients:
    print(f"{ingredient.strip()}: {count}")
```

#### Are there correlations between preparation time and recipe ratings?

```python
import re
def extract_time(process_step):
    if isinstance(process_step, list):
        process_step = ' '.join(process_step)
    
    time_pattern = r'\b\d+\s*(?:minutes?|mins?|hours?|hrs?|days?)\b'
    match = re.search(time_pattern, process_step, re.IGNORECASE)
    return match.group(0) if match else None

result['Time'] = result['directions'].apply(extract_time)

print(result[['directions', 'Time']])

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=result['Time'])
plt.title('Boxplot of Time Column')
plt.xlabel('Time')
plt.show()
```

```python
result = result.dropna()
```

```python
result.isnull().sum()
```

```python
def convert_to_minutes(time_str):
    match = re.match(r'(\d+)\s*([a-zA-Z]+)', time_str)
    if match:
        value = int(match.group(1))
        unit = match.group(2)
        
        if 'day' in unit:
            return value * 24 * 60  # 1 day = 24 * 60 minutes
        elif 'hour' in unit:
            return value * 60  # 1 hour = 60 minutes
        elif 'minute' in unit:
            return value  # minutes remain the same
    return None 
result['Time_in_minutes'] = result['Time'].apply(convert_to_minutes)

print(result[['Time', 'Time_in_minutes']])
```

```python
result = result.drop(columns=['Time'])
```

### Are there correlations between preparation time and recipe ratings?

```python
sns.scatterplot(data=result, x='Time_in_minutes', y='rating')
plt.title('Preparation Time vs Recipe Ratings')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Recipe Ratings')
plt.show()
```

```python
line_data = result.groupby('Time_in_minutes')['rating'].mean().reset_index()

plt.figure(figsize=(24, 6))
sns.lineplot(data=line_data, x='Time_in_minutes', y='rating', marker='o', color='blue')
plt.title('Average Recipe Ratings by Preparation Time (in minutes)')
plt.xlabel('Preparation Time (minutes)')
plt.ylabel('Average Recipe Ratings')
plt.grid()
plt.show()
```
